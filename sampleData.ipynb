```python
#the sin plot
Xstart = 0
Xstop = 2*np.pi
increment = 0.1
x = np.arange(Xstart,Xstop,increment)
y = np.sin(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
Xstart = 0
Xstop = 2*np.pi
increment = 0.1
x = np.arange(Xstart,Xstop,increment)
y = np.sin(x)
z = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x, y, 'g')
plt.title('sin')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid()
plt.show()



plt.subplot(2,1,2)
plt.plot(x, z, 'r')
plt.title('cos')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid()
plt.show()
```

```python
#ceating a mershgrid for x andd y values
X =  np.linspace(-5, 5, 100)
Y =  np.linspace(-5, 5, 100)
X,Y = np.meshgrid(X,Y)
```

```python
#Defining a function to calculate the Z values(height) based on X and Y
def f(X,Y):
    return np.sin(np.sqrt(X**2 + Y**2))
#calculating the values of Z for the meshgrid

Z = f(X,Y)
```

```python
#creating a three dimention contour plot
fig = plt.figure(figsize = (8,6))
ax = fig.add_subplot(111, projection='3d')
contour = ax.contour3D(X,Y,Z, 50, cmap= 'gnuplot2')
#labels and colorbars
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
fig.colorbar(contour, ax=ax, label='Z-values')
```
