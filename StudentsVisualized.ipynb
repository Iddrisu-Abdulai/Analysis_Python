```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students = pd.read_csv('C:/Users/hp/OneDrive/Desktop/StudentsPerformance.csv')
```

```python
students.head(6)
```

```python
##taking a look at the data structure
students.info()
```

```python
#Visualizing the distribution maths scores of students by gender and coloring by test preparation course
#with a boxplot
plt.figure(figsize=(10,5))
plt.title('Math score seggregated on Gender and Preparation')
sns.boxplot(data=students, y='gender',x='math score', hue='test preparation course',
            palette='cool', orient='h')
plt.show()
```

```python
#It can be seen from the graph that a lot of the student who didnt compele their test preparration
#cousre, depicted by the blue color has dominated those who did, with the females being the most
#amongs them. With respect to those who completed, has their averge math scores to be  much higher
#than the later, with males appearing top with higher math scores on top of the females.
#Males also bserved a much wide range of scores in the maths paper.
```

```python
#Visualizing the distribution maths scores of students by gender and coloring by parents level 
#of education with a bswarmplot
plt.figure(figsize=(10,5))
plt.title('Math score by Gender and Parental level of education')
sns.swarmplot(data=students, x='math score', y='gender', hue='parental level of education',
              palette='tab10')
plt.legend(loc='lower left')
plt.show()
```

```python
#it appears that some students whose parents has their level of esucation at 'some high school, high school and some colleage,
#appears to be falling below the average math score,having students with parents at 'some high school levels' being the dominant.
#While parents with 'associate's degree, master's degree and bachelor's degree', being clustered on
#and above the average math score for the bothe gender.
```

```python
#joint plot of students reading score, math score and color by gender
sns.jointplot(data=students, x='reading score', y='math score', kind='scatter', hue='gender',
              palette='winter', alpha=0.5)
plt.show()
```

```python
#from the plot it can be seen that while there are plenty females students that male, the male students 
#appears to top in both 'math and reading' scores respectively
```

```python
#a paire plot of students performances by gender
sns.pairplot(data=students, hue='gender',palette='seismic')
plt.show()
             
```
